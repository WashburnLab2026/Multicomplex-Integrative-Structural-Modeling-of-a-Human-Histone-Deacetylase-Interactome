```python
""" 
Proposed by Jules Nde, PhD, Washburn Lab, Cancer Biology, University of Kansas Medical Center (25/8/2025)

Author: Jules Nde, Washburn Lab, Cancer Biology, University of Kansas Medical Center
Copyright © 2025 by Jules Nde. All rights reserved. No part of this work may be reproduced or transmitted in any form without permission.

This script extracts one or more specified chains from a given PDB file and writes the selected chains to a new PDB file.
$python extract_chains.py input.pdb output.pdb A B  (if using the python version instead of jupyter notebook)
"""

from Bio import PDB
import os

def extract_chains(input_pdb, output_pdb, chain_ids):
    if input_pdb.endswith('.cif'):
        parser = PDB.MMCIFParser(QUIET=True)  
    elif input_pdb.endswith('.pdb'):
        parser = PDB.PDBParser(QUIET=True)  
    else:
        raise ValueError(f"Unsupported file format for {input_pdb}. Must be .pdb or .cif.")
    
    structure = parser.get_structure("structure", input_pdb)

    io = PDB.PDBIO()
    
    class ChainSelect(PDB.Select):
        def accept_chain(self, chain):
            return chain.id in chain_ids

    io.set_structure(structure)
    io.save(output_pdb, select=ChainSelect())
    print(f"Chains {', '.join(chain_ids)} extracted to {output_pdb}")

input_pdb = "../fold_2sin3_2ume1_dep1_sds3_sap30/fold_2sin3_2ume1_dep1_sds3_sap30_model_4.cif"         # Path to your input PDB file
output_pdb = "../fold_2sin3_2ume1_dep1_sds3_sap30/Sin3_Ume1_Dep1_Sds3.pdb"  # Path to save the extracted chains
#chain_ids = ["A", "B", "C", "D", "F", "I"]            # List of chain IDs to extract
chain_ids = ["A", "C", "E", "F"]

if not os.path.exists(input_pdb):
    print(f"Error: File {input_pdb} does not exist.")
else:
    extract_chains(input_pdb, output_pdb, chain_ids)
```
