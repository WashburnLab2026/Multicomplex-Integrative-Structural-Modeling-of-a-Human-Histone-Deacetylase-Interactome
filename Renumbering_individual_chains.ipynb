```python
""" 
Proposed by Jules Nde, PhD, Washburn Lab, Cancer Biology, University of Kansas Medical Center (25/8/2025)

Author: Jules Nde, Washburn Lab, Cancer Biology, University of Kansas Medical Center
Copyright © 2025 by Jules Nde. All rights reserved. No part of this work may be reproduced or transmitted in any form without permission.

This script renumbers the residues of specific chains (e.g., starting from residue number 1) in a given PDB 
and saves the modified structure into a new PDB file. 

Run the script using the command $python renumber_chains.py input.pdb output.pdb A B; (if using the python version instead of jupyter notebook)
This example renumbers residues of chains A and B, starting from residue number 1.
"""

from Bio.PDB import PDBParser, PDBIO, Select

class ChainRenumberingSelect(Select):
    def __init__(self, target_chains):
        self.target_chains = target_chains

    def accept_residue(self, residue):
        return True

    def accept_atom(self, atom):
        return True

# Function to renumber specific chains
def renumber_chains(structure, target_chains):
    for model in structure:
        for chain in model:
            if chain.id in target_chains:
                new_resseq = 937
                new_residues = []
                for res in list(chain):
                    hetfield, _, icode = res.id
                    res.id = (hetfield, new_resseq, icode)
                    new_residues.append(res)
                    new_resseq += 1
                for res in list(chain):
                    chain.detach_child(res.id)
                for res in new_residues:
                    chain.add(res)
    return structure

# Main function to load, process, and save
def renumber_pdb(input_pdb_path, output_pdb_path, chains_to_renumber):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure("structure", input_pdb_path)

    renumbered_structure = renumber_chains(structure, chains_to_renumber)

    io = PDBIO()
    io.set_structure(renumbered_structure)
    io.save(output_pdb_path, select=ChainRenumberingSelect(chains_to_renumber))

    print(f"Chains {', '.join(chains_to_renumber)} renumbered and saved to {output_pdb_path}")

# Example usage (update paths and chains as needed)
input_pdb = "cluster1_1_chains_ren.pdb"        # Replace with your file
output_pdb = "cluster1_1_chains_renumb.pdb"      # Output file
#chains_to_fix = ["B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N"]             # Chains to renumber
chains_to_fix = ["L"]

renumber_pdb(input_pdb, output_pdb, chains_to_fix)
```
