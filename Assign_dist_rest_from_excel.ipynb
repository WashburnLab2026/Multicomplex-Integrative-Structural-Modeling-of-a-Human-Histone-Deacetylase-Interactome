This scripts are used to assign distance restraints from a given excel crosslinking table by:  

1-) Separate a given excel file into INTRA- and INTER-crosslinks. 

2-) Create a tabulated file with fixed target distance restraints, which is used as an input file to guide the docking with HADDOCK. 

3-) Create a tabulated file with variable target distance restraints, which is used as an input file to guide the docking with HADDOCK. 


```python
# Separate a given excel file into INTRA- and INTER-crosslinks

import pandas as pd

# Replace with your actual file path
input_file = 'Input_contacts.xlsx'

# Read the Excel file
df = pd.read_excel(input_file, header=None)

# Check if columns 0 and 2 are the same or different
same_df = df[df[0] == df[2]]
inter_df = df[df[0] != df[2]]

# Save results to two separate Excel files
same_df.to_excel('contacts_docking_intra.xlsx', index=False, header=False)
inter_df.to_excel('contacts_for_docking_inter.xlsx', index=False, header=False)

print("Saved 'same.xlsx' and 'inter.xlsx' successfully.")
```

```python
"""
Create a tabulated file with fixed target distance restraints, 
which is used as an input file to guide the docking with HADDOCK
"""
import pandas as pd

# === Step 1: Load Excel file ===
# Replace 'your_file.xlsx' with the actual filename
input_file = '8GHN_A_B_C_inter_contacts.xlsx'

# Load the Excel file (assumes no header)
df = pd.read_excel(input_file) #, header=None)

# Assign column names for clarity
#df.columns = ['segid1', 'resid1', 'segid2', 'resid2']
df.columns = ['segid1', 'resid1', 'segid2', 'resid2', 'distance']

# === Step 2: Create formatted restraints ===
restraints = []

for _, row in df.iterrows():
    line = (
        f"assi ( segid {row.segid1} and resid {row.resid1} and name CA ) "
        f"( segid {row.segid2} and resid {row.resid2} and name CA ) 25 25 0"
    )
    restraints.append(line)

# === Step 3: Save to .tbl file ===
output_file = '8GHN_distance_restraints_fixed_25A.tbl'

with open(output_file, 'w') as f:
    for line in restraints:
        f.write(line + '\n')

print(f"✅ Restraints saved to '{output_file}'")
```

```python
"""
Create a tabulated file with variable target distance restraints, 
which is used as an input file to guide the docking with HADDOCK
"""

import pandas as pd

# === Step 1: Load Excel file ===
input_file = '8GHN_A_B_C_inter_contacts.xlsx'  # Replace with your actual filename
df = pd.read_excel(input_file) #, header=None)

# Assign column names
df.columns = ['segid1', 'resid1', 'segid2', 'resid2', 'distance']

# === Step 2: Create formatted restraints ===
restraints = []

for _, row in df.iterrows():
    dist = float(row.distance)
    # You can adjust bounds if needed; here we set upper and lower equal to actual distance
    line = (
        f"assi ( segid {row.segid1} and resid {row.resid1} and name CA ) "
        f"( segid {row.segid2} and resid {row.resid2} and name CA ) "
        f"{dist:.2f} {dist:.2f} 0"
    )
    restraints.append(line)

# === Step 3: Save to .tbl file ===
output_file = '8GHN_distance_restraints_variable_dist.tbl'

with open(output_file, 'w') as f:
    for line in restraints:
        f.write(line + '\n')

print(f"✅ Restraints saved to '{output_file}' with custom distances.")
```
